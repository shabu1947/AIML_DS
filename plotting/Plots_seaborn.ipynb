```
# Distribution of datasets
```

```
import seaborn as sns

%matplotlib inline
```

```
# Data to plot, we are getting from seaborn's built in datasets.

# Here I am using tips data set

tips_ds = sns.load_dataset('tips')
tips_ds.head()
```

```
# Uni-variate distribution. In this we will check single column data, how it is varying.
# for this I will use distplot, distribution

sns.distplot(tips_ds['total_bill'])
```

```
sns.distplot(tips_ds['total_bill'], kde=False, bins=30)
```

```
sns.kdeplot(tips_ds['total_bill'])
```

```
# Joint plot

sns.jointplot(x='total_bill', y='tip', data=tips_ds)
```

```
sns.jointplot(x='total_bill', y='tip', data=tips_ds, kind='hex')
```

```
# Regressions in jointplot
sns.jointplot(x='total_bill', y='tip', data=tips_ds, kind='reg')
```

```
# kde
sns.jointplot(x='total_bill', y='tip', data=tips_ds, kind='kde')
```

```
# From the above joint plots we can analyse more based on the color densitiy, the more the color density more the samples at that point.
```

```
# Next one is pair plots
# pairplot gives comparision between all numerical columns in the data structure

sns.pairplot(tips_ds)
```

```
# we can do compairison along with categorical values, by passing on hue option
# I am going to plot wrt to sex column in the data structure.
sns.pairplot(tips_ds, hue='sex')
```

```
# Categorical plots
```

```
# Here I am using bar plot to plot categorical data

# x --> categorical column
# y --> numerical column
# below plot shows the average total bill for male and females
# by default estimator is mean
sns.barplot(x='sex', y='total_bill', data=tips_ds)
```

```
# changing the estimator from default mean to someother function,here I am using std.

import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips_ds, estimator=np.std)
```

```
# Count plot, which similar to pands .count for tips dataframe against gender column
# it will show the count of occurunces of categorical data

sns.countplot(x='sex',data=tips_ds)
```

```
# Box plot is often used in explanatory data analysis. This type of graph is
# used to show the shape of the distribution, its central value, and its variability. 
# In a box and whisker plot: the ends of the box are the upper and lower quartiles,
# so the box spans the interquartile range
```

```
sns.boxplot(x='day',y='total_bill',data=tips_ds)
```

```
sns.boxplot(x='day',y='total_bill',data=tips_ds, hue='sex')
```
