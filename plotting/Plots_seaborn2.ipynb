```
import seaborn as sns

%matplotlib inline
```

```
# loading tips data set from the seaborn in built datasets
tips = sns.load_dataset('tips')
tips.head()
```

```
# Loading flights data set from seaborn inbuilt dataset

flights = sns.load_dataset('flights')
flights.head()
```

```
# to plot heat map, we should have data formatted. We can use pivot table / correlation
# here I will go with correlation.
# with pivot or correlation, we can get columns names in rows as well, creating a relation between each.
# as shown below

tc = tips.corr()
tc
```

```
sns.heatmap(tc, annot=True)
```

```
# Now here I am using pivot on flights data

fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

```
sns.heatmap(fp, linecolor='white', linewidths=1)
```

```
# Cluster maps.
# Cluster map clusters similar data and creates a heat map based on the clusturred data

sns.clustermap(fp)
```
